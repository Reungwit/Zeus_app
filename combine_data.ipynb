```python
import pandas as pd

file_old = 'zeus_dataset_raw.csv'
file_new = 'zeus_dataset_raw_newdata.csv'
output_raw = 'zeus_master_raw.csv'

print("📥 กำลังโหลดข้อมูลดิบทั้ง 2 ชุด...")
df_old = pd.read_csv(file_old)
df_new = pd.read_csv(file_new)

print(f"   - ไฟล์เก่า: {len(df_old)} แถว")
print(f"   - ไฟล์ใหม่: {len(df_new)} แถว")

# รวมร่างข้อมูลแบบดิบๆ (ยังไม่ล้าง)
df_master_raw = pd.concat([df_old, df_new], ignore_index=True)

# เซฟเป็นไฟล์ Master Raw
df_master_raw.to_csv(output_raw, index=False)
print("-" * 40)
print(f"✅ รวมไฟล์ดิบสำเร็จ! ได้ไฟล์ {output_raw} จำนวนรวม {len(df_master_raw)} แถว")
```

```python
import pandas as pd
from datetime import timedelta

input_file = 'zeus_master_cleaned.csv'
train_file = 'zeus_train_final.csv'       # ข้อมูลสำหรับให้ AI เรียน
test_file = 'zeus_test_unseen.csv'        # ข้อมูลสำหรับใช้สอบ (7 วันสุดท้าย)

print(f"📥 1. กำลังโหลดข้อมูลสะอาดจาก: {input_file}")
df = pd.read_csv(input_file)

# แปลงคอลัมน์ datetime ให้เป็นรูปแบบเวลาของ Pandas
df['datetime'] = pd.to_datetime(df['datetime'])

# หาวันที่ล่าสุดที่มีใน Dataset
max_date = df['datetime'].max()
# คำนวณจุดตัด (Cut-off Date) ย้อนกลับไป 7 วัน
cutoff_date = max_date - timedelta(days=7)

print(f"   📅 วันที่ข้อมูลเก่าสุด: {df['datetime'].min()}")
print(f"   📅 วันที่ข้อมูลใหม่สุด: {max_date}")
print(f"   ✂️ จุดตัดข้อมูล (Cut-off): {cutoff_date}")

# ทำการแบ่งข้อมูล
print("\n🪓 2. กำลังแบ่งข้อมูล (Train / Test Split)...")
df_train = df[df['datetime'] <= cutoff_date]
df_test = df[df['datetime'] > cutoff_date]

# บันทึกเป็นไฟล์ CSV
df_train.to_csv(train_file, index=False)
df_test.to_csv(test_file, index=False)

print("\n✅ แบ่งข้อมูลสำเร็จเรียบร้อย!")
print(f"   📚 ข้อมูลสำหรับเทรน (Train): {len(df_train)} แถว -> บันทึกเป็น {train_file}")
print(f"   📝 ข้อมูลข้อสอบ (Test 7 วัน): {len(df_test)} แถว -> บันทึกเป็น {test_file}")
```
